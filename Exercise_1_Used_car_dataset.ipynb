```python
# import libraries
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd

tf.random.set_seed(42)
# Find suitable dataset
used_cars_df = pd.read_csv("https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/usedcars.csv")
```

```python
used_cars_df
```

```python
X = used_cars_df.drop("price",axis=1)
y = used_cars_df.price
```

```python
X
```

```python
y
```

```python

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1,random_state=42)

# for missing columns
# Assuming X_train and X_test have the same columns
common_columns = X_train.columns

# Reorder columns in X_test to match the order in X_train
X_test = X_test[common_columns]
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
x_model_dummies = pd.get_dummies(X_train.model)
x_color_dummmies = pd.get_dummies(X_train.color)
x_transmission_dummies = pd.get_dummies(X_train.transmission)
```

```python
X_train = pd.concat([X_train,x_model_dummies],axis=1)
X_train = X_train.drop("model",axis=1)
```

```python
X_train = pd.concat([X_train,x_color_dummmies],axis=1)
X_train = X_train.drop("color",axis=1)
```

```python
X_train = pd.concat([X_train,x_transmission_dummies],axis=1)
X_train = X_train.drop("transmission",axis=1)
```

```python
X_train
```

```python
x_model_dummies = pd.get_dummies(X_test.model)
x_color_dummmies = pd.get_dummies(X_test.color)
x_transmission_dummies = pd.get_dummies(X_test.transmission)
```

```python
X_test = pd.concat([X_test,x_model_dummies],axis=1)
X_test = X_test.drop("model",axis=1)
```

```python
X_test = pd.concat([X_test,x_color_dummmies],axis=1)
X_test = X_test.drop("color",axis=1)
```

```python
X_test = pd.concat([X_test,x_transmission_dummies],axis=1)
X_test = X_test.drop("transmission",axis=1)
```

```python
X_test
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1),
    tf.keras.layers.Dense(10,activation="relu"),
    tf.keras.layers.Dense(1),
    ])
model.compile(loss=tf.keras.losses.mae,optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),metrics=["mae"])
model.fit(X_train,y_train,epochs=200)
```

```python
# code courtesy of chatgpt
def add_missing_columns(df1, df2):
    missing_columns = set(df2.columns) - set(df1.columns)
    for col in missing_columns:
        df1[col] = 0  # You can change 'None' to any default value you want
add_missing_columns(X_test,X_train)
```

## Testing normalization

```python
y_train_max_values = tf.math.reduce_max(y_train, axis=0, keepdims=True)
y_train_norm = y_train / tf.cast(y_train_max_values, y_test.dtype)
```

```python
y_train_max_values
```

```python
y_test_max_values = tf.math.reduce_max(y_test, axis=0, keepdims=True)
y_test_norm = y_test / tf.cast(y_test_max_values, y_test.dtype)
```

```python
y_test_max_values
```

```python
y_test_norm
```

```python
norm_model = tf.keras.Sequential([
    tf.keras.layers.Dense(1),
    tf.keras.layers.Dense(12,activation="relu"),
    tf.keras.layers.Dense(1),
    ])
norm_model.compile(loss=tf.keras.losses.mae,optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),metrics=["mae"])
norm_model.fit(X_train,y_train_norm,epochs=200)
```

```python
norm_model.evaluate(X_test,y_test_norm)
```
